### 2. Violation Analysis

```python
# 2.1 Violation type distribution and Top 10 violations

# Assuming 'df_violations_exploded' already exists from previous steps
# If not, ensure the 'violation_type' column is processed into a list and exploded

# Calculate the distribution of all individual violation types
violation_distribution = df_violations_exploded['violation_type_list'].value_counts(normalize=True) * 100

print("\nOverall Violation Type Distribution (%):")
print(violation_distribution.round(2))

print("\nTop 10 Most Common Violations:")
display(violation_distribution.head(10).round(2))
```

```python
# 2.2 Parking vs non-parking violations

# Define keywords to identify parking violations
parking_keywords = ['PARKING', 'NO PARKING', 'WRONG PARKING', 'FOOTPATH PARKING', 'MAIN ROAD']

# Function to categorize a violation as parking or non-parking
def categorize_violation(violation):
    for keyword in parking_keywords:
        if keyword in violation.upper():
            return 'Parking Violation'
    return 'Non-Parking Violation'

# Apply the categorization
df_violations_exploded['violation_category'] = df_violations_exploded['violation_type_list'].apply(categorize_violation)

# Calculate the distribution of parking vs non-parking violations
parking_non_parking_distribution = df_violations_exploded['violation_category'].value_counts(normalize=True) * 100

print("\nParking vs Non-Parking Violations Distribution (%):")
display(parking_non_parking_distribution.round(2))

# Plotting the distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=parking_non_parking_distribution.index, y=parking_non_parking_distribution.values, palette='coolwarm')
plt.title('Distribution of Parking vs Non-Parking Violations')
plt.ylabel('Percentage (%)')
plt.xlabel('Violation Category')
plt.show()
```

```python
# 3.5 Commercial vs private vehicle violations

# Define keywords for commercial vehicle types (this might need adjustment based on specific data)
commercial_keywords = ['LORRY', 'TRUCK', 'BUS', 'TEMPO', 'GOODS', 'MAXI-CAB', 'TANKER', 'LGV', 'HGV', 'AUTO']

def categorize_vehicle_commercial_private(vehicle_type):
    if any(keyword in str(vehicle_type).upper() for keyword in commercial_keywords):
        return 'Commercial'
    elif pd.notnull(vehicle_type):
        return 'Private'
    return 'Unknown'

# Apply categorization
df['vehicle_category'] = df['vehicle_type'].apply(categorize_vehicle_commercial_private)

# Calculate violations by vehicle category
commercial_private_violations = df['vehicle_category'].value_counts(normalize=True) * 100

print("\nCommercial vs Private Vehicle Violations Distribution (%):")
display(commercial_private_violations.round(2))

# Plotting the distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=commercial_private_violations.index, y=commercial_private_violations.values, palette='viridis')
plt.title('Distribution of Commercial vs Private Vehicle Violations')
plt.ylabel('Percentage (%)')
plt.xlabel('Vehicle Category')
plt.show()
```

### 5. Temporal Analysis

#### Monthly

```python
# 5.1 Violations by month and month-over-month growth

# Extract month from 'created_datetime'
df['month'] = df['created_datetime'].dt.to_period('M')

monthly_violations = df['month'].value_counts().sort_index()

print("\nViolations by Month:")
display(monthly_violations)

# Calculate month-over-month growth
monthly_growth = monthly_violations.pct_change() * 100

print("\nMonth-over-Month Growth in Violations (%):")
display(monthly_growth.round(2))

# Plotting monthly violations
plt.figure(figsize=(12, 6))
monthly_violations.plot(kind='bar', color='skyblue')
plt.title('Monthly Violation Trends')
plt.xlabel('Month')
plt.ylabel('Number of Violations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5.2 Violations by weekday and weekday vs weekend

# Extract weekday from 'created_datetime'
df['weekday'] = df['created_datetime'].dt.day_name()

violations_by_weekday = df['weekday'].value_counts().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

print("\nViolations by Weekday:")
display(violations_by_weekday)

# Categorize into weekday vs. weekend
df['day_type'] = df['weekday'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
weekend_weekday_violations = df['day_type'].value_counts(normalize=True) * 100

print("\nWeekday vs Weekend Violations Distribution (%):")
display(weekend_weekday_violations.round(2))

# Plotting violations by weekday
plt.figure(figsize=(10, 6))
sns.barplot(x=violations_by_weekday.index, y=violations_by_weekday.values, palette='viridis')
plt.title('Violations by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Violations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Hourly

```python
# 5.3 Peak violation hours, morning vs evening patterns, rush hour effects

# Extract hour from 'created_datetime'
df['hour'] = df['created_datetime'].dt.hour

hourly_violations = df['hour'].value_counts().sort_index()

print("\nHourly Violation Counts:")
display(hourly_violations)

# Identify peak hours
peak_hours = hourly_violations.nlargest(3)
print("\nTop 3 Peak Violation Hours:")
display(peak_hours)

# Morning vs Evening Patterns (e.g., 6 AM - 12 PM as Morning, 4 PM - 8 PM as Evening rush)
def time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning (6AM-12PM)'
    elif 16 <= hour < 20:
        return 'Evening Rush (4PM-8PM)'
    else:
        return 'Other'

df['time_period'] = df['hour'].apply(time_of_day)
time_period_violations = df['time_period'].value_counts()

print("\nViolations by Time Period (Morning/Evening Rush):")
display(time_period_violations)

# Plotting hourly violations
plt.figure(figsize=(12, 6))
hourly_violations.plot(kind='line', marker='o', color='coral')
plt.title('Hourly Violation Trends')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Violations')
plt.xticks(range(0, 24))
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 6. Validation Statistics

```python
# 6.1 Extracting Validation Metrics: Approval rate, Rejection rate, Processing backlog, SCITA transmission success

# Ensure 'validation_status' is handled correctly (e.g., fillna for pending/null values)
# For 'validation_status', let's assume NaN means 'pending'
df['validation_status_filled'] = df['validation_status'].fillna('pending')

total_violations = len(df)

# Approval Rate
approved_count = df['validation_status_filled'].value_counts().get('approved', 0)
approval_rate = (approved_count / total_violations) * 100

# Rejection Rate
rejected_count = df['validation_status_filled'].value_counts().get('rejected', 0)
rejection_rate = (rejected_count / total_violations) * 100

# Processing Backlog (Pending)
pending_count = df['validation_status_filled'].value_counts().get('pending', 0)
processing_backlog = (pending_count / total_violations) * 100

# SCITA Transmission Success
# Assuming 'data_sent_to_scita' is True for success, False otherwise
scita_sent_count = df['data_sent_to_scita'].sum()
scita_transmission_success = (scita_sent_count / total_violations) * 100

print(f"Approval Rate: {approval_rate:.2f}%")
print(f"Rejection Rate: {rejection_rate:.2f}%")
print(f"Processing Backlog (Pending): {processing_backlog:.2f}%")
print(f"SCITA Transmission Success: {scita_transmission_success:.2f}%")

# Optional: Display counts for each validation status
print("\nValidation Status Counts:")
display(df['validation_status_filled'].value_counts())

# Plotting Validation Status Distribution
validation_status_counts = df['validation_status_filled'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=validation_status_counts.index, y=validation_status_counts.values, palette='pastel')
plt.title('Distribution of Validation Status')
plt.xlabel('Validation Status')
plt.ylabel('Number of Violations')
plt.show()
```

### 7. Congestion Indicators

```python
# 7.1 Create Congestion Score per junction

# Filter for relevant parking violation types that indicate congestion
congestion_parking_violations = [
    'WRONG PARKING',
    'NO PARKING',
    'FOOTPATH PARKING',
    'PARKING IN A MAIN ROAD'
]

# Filter the exploded DataFrame to only include these congestion-related parking violations
df_congestion = df_violations_exploded[df_violations_exploded['violation_type_list'].isin(congestion_parking_violations)].copy()

# Calculate counts for each congestion-related violation per junction
junction_congestion_counts = df_congestion.groupby(['junction_name', 'violation_type_list']).size().unstack(fill_value=0)

# Ensure all expected columns exist, fill with 0 if not
for col in ['WRONG PARKING', 'NO PARKING', 'FOOTPATH PARKING', 'PARKING IN A MAIN ROAD']:
    if col not in junction_congestion_counts.columns:
        junction_congestion_counts[col] = 0

# Apply the weighted formula to calculate Congestion Score
junction_congestion_counts['Congestion Score'] = (
    0.4 * junction_congestion_counts['WRONG PARKING'] +
    0.3 * junction_congestion_counts['NO PARKING'] +
    0.2 * junction_congestion_counts['FOOTPATH PARKING'] +
    0.1 * junction_congestion_counts['PARKING IN A MAIN ROAD']
)

# Rank junctions by Congestion Score
ranked_junctions = junction_congestion_counts.sort_values(by='Congestion Score', ascending=False)

print("\nJunctions Ranked by Congestion Score (Top 10):")
display(ranked_junctions[['Congestion Score']].head(10))

# Plotting the top 10 congested junctions
plt.figure(figsize=(12, 7))
sns.barplot(x=ranked_junctions['Congestion Score'].head(10), y=ranked_junctions.head(10).index, palette='Reds_d')
plt.title('Top 10 Junctions by Congestion Score')
plt.xlabel('Congestion Score')
plt.ylabel('Junction Name')
plt.show()
```

### 9. Enforcement Efficiency

```python
# 9.1 Violations detected, approved, rejected, and approval percentage by police station

# Group by 'police_station' and 'validation_status_filled' to get counts
enforcement_stats = df.groupby(['police_station', 'validation_status_filled']).size().unstack(fill_value=0)

# Calculate total violations detected (sum of all statuses)
enforcement_stats['Violations Detected'] = enforcement_stats.sum(axis=1)

# Get approved and rejected counts
enforcement_stats['Violations Approved'] = enforcement_stats.get('approved', 0)
enforcement_stats['Violations Rejected'] = enforcement_stats.get('rejected', 0)

# Calculate Approval Percentage
enforcement_stats['Approval Percentage'] = (
    enforcement_stats['Violations Approved'] / enforcement_stats['Violations Detected']
) * 100

# Select relevant columns for display and sort by 'Violations Detected'
enforcement_efficiency = enforcement_stats[[
    'Violations Detected',
    'Violations Approved',
    'Violations Rejected',
    'Approval Percentage'
]].sort_values(by='Violations Detected', ascending=False)

print("\nEnforcement Efficiency by Police Station (Top 10 by Violations Detected):")
display(enforcement_efficiency.head(10))

# Plotting Approval Percentage for top 10 stations by violations detected
plt.figure(figsize=(12, 7))
sns.barplot(x=enforcement_efficiency['Approval Percentage'].head(10), y=enforcement_efficiency.head(10).index, palette='viridis')
plt.title('Approval Percentage by Police Station (Top 10 by Violations Detected)')
plt.xlabel('Approval Percentage (%)')
plt.ylabel('Police Station')
plt.xlim(0, 100)
plt.show()
```

```python
# Analyzing specific parking violations vs high-traffic junctions
parking_violations = [
    '["WRONG PARKING"]',
    '["NO PARKING"]',
    '["PARKING IN A MAIN ROAD"]',
    '["PARKING NEAR ROAD CROSSING"]'
]

# Filter for relevant violations and top 15 junctions (as a proxy for congestion points)
top_junctions = df['junction_name'].value_counts().nlargest(15).index
filtered_df = df[df['violation_type'].isin(parking_violations) & df['junction_name'].isin(top_junctions)]

# Create a pivot table for the heatmap
parking_heatmap_data = pd.crosstab(filtered_df['junction_name'], filtered_df['violation_type'])

plt.figure(figsize=(14, 10))
sns.heatmap(parking_heatmap_data, annot=True, fmt='d', cmap='Reds')
plt.title('Heatmap: Parking Violations at Major Junctions (Congestion Proxy)')
plt.xlabel('Violation Type')
plt.ylabel('Junction Name')
plt.show()
```

```python
# Defining parking violations that directly impact traffic flow (congestion)
congestion_causing_violations = [
    '["WRONG PARKING"]',
    '["NO PARKING"]',
    '["PARKING IN A MAIN ROAD"]',
    '["PARKING NEAR ROAD CROSSING"]',
    '["FOOTPATH PARKING"]'
]

# Identifying the top 20 junctions with the most violations to represent high-congestion areas
top_congestion_junctions = df[df['junction_name'] != 'No Junction']['junction_name'].value_counts().nlargest(20).index

# Filtering data
congestion_df = df[
    df['violation_type'].isin(congestion_causing_violations) &
    df['junction_name'].isin(top_congestion_junctions)
]

# Creating the cross-tabulation for the heatmap
heatmap_data = pd.crosstab(congestion_df['junction_name'], congestion_df['violation_type'])

# Plotting the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlOrRd', cbar_kws={'label': 'Number of Violations'})
plt.title('Heatmap: Parking Violations vs. Likely Congestion Impact (Top Junctions)', fontsize=16)
plt.xlabel('Parking Violation Type', fontsize=12)
plt.ylabel('Junction (High Traffic Areas)', fontsize=12)
plt.show()
```

4. **Congestion Correlation**:
